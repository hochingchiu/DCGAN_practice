```python
data_dir = '../../datasets/CelebA'

time_run = datetime.now().strftime('%Y%m%d_%H%M%S')
save_dir = data_dir + '/results/' + time_run
os.makedirs(save_dir, exist_ok=True)
```

```python
##Run this cell if transform mean and std are unknown

get_stat_transform = transforms.Compose([
transforms.ToTensor()
])

get_stat_dataset = datasets.ImageFolder(data_dir, get_stat_transform)

def get_transform_stats(dataset):
    """
    Implemented running average for std for multiple batches, i.e. batch_size != no. of image in dataset 
    (it might be necessary when dataset is extremely huge and there is not sufficient memory),
    with the trade off of not getting true std for the whole data set (average of std of batches != std of whole data set)
    However, I think it is better than nothing.
    """
    dataloader = torch.utils.data.DataLoader(dataset, batch_size=128, shuffle=False, num_workers=7)
    dataiter = iter(dataloader)
    running_mean = torch.zeros(3)
    running_std = torch.zeros(3)
        
    for i in range(len(dataloader)):
        n = i + 1
        data, _ = dataiter.next()

        batch_mean = torch.from_numpy(np.mean(data.numpy(), axis=(0, 2, 3)))
        batch_std = torch.from_numpy(np.std(data.numpy(), axis=(0, 2, 3)))

        delta_mean = batch_mean - running_mean
        running_mean += delta_mean / n
        delta_std = batch_std - running_std
        running_std += delta_std / n
    print('mean:', running_mean, 'std:', running_std)
    return np.array(running_mean), np.array(running_std)

transform_mean, transform_std = get_transform_stats(get_stat_dataset)
#transform_mean = np.array([0.4914, 0.4822, 0.4465])
#transform_std = np.array([0.2470, 0.2435, 0.2616])
```

```python
# Number of workers for dataloader
workers = 7

# Batch size during training
batch_size = 128

# Spatial size of training images. All images will be resized to this
#   size using a transformer.
image_size = 64

# Number of channels in the training images. For color images this is 3
nc = 3

# Size of z latent vector (i.e. size of generator input)
nz = 100

# Size of feature maps in generator
nfm = 64

# Size of feature maps in discriminator
ndf = 64

# Number of training epochs
num_epochs = 5

# Learning rate for optimizers
lr = 0.0002

# Beta1 hyperparam for Adam optimizers
beta1 = 0.5

# Number of GPUs available. Use 0 for CPU mode.
ngpu = 1

seed = 999
np.random.seed(seed)
random.seed(seed)
torch.manual_seed(seed)
```

```python
if transform_mean.any() == None:
    transform_mean = np.array([0.5061, 0.4254, 0.3828])
if transform_std.any() == None:
    transform_std = np.array([0.3103, 0.2899, 0.2893])
    
transform = transforms.Compose([
    transforms.Resize(image_size),
    transforms.CenterCrop(image_size),
#    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize(mean=transform_mean, std=transform_std)
])

dataset = datasets.ImageFolder(data_dir, transform)
data_loader = torch.utils.data.DataLoader(dataset, 
                                          batch_size=batch_size,
                                          shuffle=True,
                                          num_workers=workers
                                         )

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

real_batch = next(iter(data_loader))
plt.figure(figsize=(8,8))
plt.axis("off")
plt.title("Training Images")
plt.imshow(np.transpose(torchvision.utils.make_grid(real_batch[0].to(device)[:64], padding=2, normalize=True).cpu(),(1,2,0)))
```

```python
# parameter initialization

def initialize_weights(module):
    if isinstance(module, nn.Conv2d):
        nn.init.normal_(module.weight.data, 0.0, 0.02)
    elif isinstance(module, nn.BatchNorm2d):
        nn.init.normal_(module.weight.data, 1.0, 0.02)
        nn.init.constant_(module.bias.data, 0)
    elif isinstance(module, nn.Linear):
        nn.init.constant_(module.bias.data, 0)
```

```python
# netG
class Generator(nn.Module):
    def __init__(self):
        super(Generator, self).__init__()
        self.main = nn.Sequential(
            #1st layer
            nn.ConvTranspose2d(nz, nfm * 16, 4, 1, 0, bias=False),
            nn.BatchNorm2d(nfm * 16),
            nn.ReLU(True),
            #2nd layer
            nn.ConvTranspose2d(nfm * 16, nfm * 8, 4, 2, 1, bias=False),
            nn.BatchNorm2d(nfm * 8),
            nn.ReLU(True),
            #3rd layer
            nn.ConvTranspose2d(nfm * 8, nfm * 4, 4, 2, 1, bias=False),
            nn.BatchNorm2d(nfm * 4),
            nn.ReLU(True),
            #4th layer
            nn.ConvTranspose2d(nfm * 4, nfm * 2, 4, 2, 1, bias=False),
            nn.BatchNorm2d(nfm * 2),
            nn.ReLU(True),
            #5th layer
            nn.ConvTranspose2d(nfm * 2, nc, 4, 2, 1, bias=False),
            nn.Tanh()
        )
        
    def forward(self, x):
        return self.main(x)
```
